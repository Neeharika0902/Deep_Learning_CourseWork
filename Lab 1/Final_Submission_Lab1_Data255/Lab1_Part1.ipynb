This cell defines the AnimeDataset class, which inherits from torch.utils.data.Dataset. This class provides a structure for organizing the input features (user and anime data) and the target ratings for each data point. It allows for efficient batching of data during model training and evaluation. The features are transformed into PyTorch tensors for compatibility with the neural network, and the dataset is prepared for feeding into a DataLoader, which will handle mini-batch training.

user_features: Consists of Days Watched, Mean Score, and Completed columns (user-related).

anime_features: Includes Score, Episodes, Rank, Popularity, and Favorites columns (anime-related).

rating: The target variable that the model will predict.

This structure provides a clear separation of user and anime features, facilitating their use in a Wide & Deep model.

```python
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from torch.utils.data import Dataset, DataLoader
import numpy as np
import matplotlib.pyplot as plt

# Define the Dataset class
class AnimeDataset(Dataset):
    def __init__(self, data):
        # Define user and anime features and target (rating)
        self.user_features = data[['Days Watched', 'Mean Score', 'Completed']].values
        self.anime_features = data[['Score', 'Episodes', 'Rank', 'Popularity', 'Favorites']].values
        self.rating = data['rating'].values
        
    def __len__(self):
        return len(self.rating)
    
    def __getitem__(self, idx):
        user_feats = torch.tensor(self.user_features[idx], dtype=torch.float32)
        anime_feats = torch.tensor(self.anime_features[idx], dtype=torch.float32)
        rating = torch.tensor(self.rating[idx], dtype=torch.float32)
        return user_feats, anime_feats, rating
```

In this cell, the Wide & Deep neural network model is defined using PyTorch. This model combines the advantages of both wide and deep learning:

Wide Part: A simple linear layer (self.wide) that directly models feature interactions for faster learning of memorization tasks. It uses a linear combination of both user and anime features, enabling the model to capture specific feature interactions.

Deep Part: A multi-layer perceptron (MLP) that learns deep patterns and generalizations from the input data. It consists of several fully connected layers:

The first layer (deep_fc1) has 256 units.

The second layer (deep_fc2) reduces this to 64 units.

The third layer (deep_fc3) further reduces the dimensionality to 32 units.

The final layer (deep_fc4) outputs a single value as the predicted rating.

Dropout regularization (self.dropout) is used after the first two layers to prevent overfitting. 

The ReLU activation function (self.relu) is applied after each hidden layer to introduce non-linearity and ensure the model can learn complex patterns.

The model outputs a combined prediction from both the wide and deep components, scaled to the range of 1 to 10 using the sigmoid activation function, which ensures the predicted rating is within the expected bounds.

```python
# Wide & Deep Model Architecture with adjusted Dropout and L2 Regularization
class WideAndDeepModel(nn.Module):
    def __init__(self, num_user_features, num_anime_features):
        super(WideAndDeepModel, self).__init__()
        
        # Wide part - Linear layer for feature interactions
        self.wide = nn.Linear(num_user_features + num_anime_features, 1)
        
        # Deep part - Fully connected layers
        self.deep_fc1 = nn.Linear(num_user_features + num_anime_features, 256)
        self.deep_fc2 = nn.Linear(256, 64)
        self.deep_fc3 = nn.Linear(64, 32)
        self.deep_fc4 = nn.Linear(32, 1)
        
        # Further tuning dropout to 0.1
        self.dropout = nn.Dropout(0.1)
        
        # Activation function
        self.relu = nn.ReLU()
    
    def forward(self, user_feats, anime_feats):
        # Wide part
        wide_input = torch.cat([user_feats, anime_feats], dim=1)
        wide_output = self.wide(wide_input)
        
        # Deep part
        deep_input = torch.cat([user_feats, anime_feats], dim=1)
        deep_output = self.relu(self.deep_fc1(deep_input))
        deep_output = self.dropout(deep_output)  # Dropout after first hidden layer
        deep_output = self.relu(self.deep_fc2(deep_output))
        deep_output = self.dropout(deep_output)  # Dropout after second hidden layer
        deep_output = self.relu(self.deep_fc3(deep_output))  # New layer added
        deep_output = self.deep_fc4(deep_output)
        
        # Combine wide and deep outputs
        combined_output = wide_output + deep_output
        
        # Scale output to [1, 10]
        final_output = torch.sigmoid(combined_output) * 9 + 1
        
        return final_output.squeeze()
```

This cell handles the preparation of the dataset and initialization of the Wide & Deep model:

Dataset Preparation: The cleaned dataset (subset_df_clean) is split into training and testing sets using an 80/20 split, which is a common practice to ensure that the model is evaluated on data it hasn’t seen during training. The AnimeDataset class defined earlier is used to create PyTorch datasets for both the training and testing sets.

DataLoader: The DataLoader is used to load the data in batches of 64 for both the training and testing sets. Batching improves training efficiency and memory usage during model training.

Model Initialization: The Wide & Deep model is initialized with 3 user features (Days Watched, Mean Score, and Completed) and 5 anime features (Score, Episodes, Rank, Popularity, Favorites).

Loss Function & Optimizer:
    
    The Mean Squared Error (MSE) loss function is used, as this is a regression problem (predicting continuous ratings). MSE measures the average squared difference between the predicted ratings and the actual ratings.
    
    The Adam optimizer is chosen with a learning rate of 0.0001, which is a common choice for deep learning models as it adjusts the learning rate during training and generally performs well across a variety of tasks.

```python
# Prepare the dataset
X_train, X_test = train_test_split(subset_df_clean, test_size=0.2, random_state=42)
train_dataset = AnimeDataset(X_train)
test_dataset = AnimeDataset(X_test)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

# Initialize the model
num_user_features = 3  # 'Days Watched', 'Mean Score', 'Completed'
num_anime_features = 5  # 'Score', 'Episodes', 'Rank', 'Popularity', 'Favorites'
model = WideAndDeepModel(num_user_features, num_anime_features)

# Define loss function and optimizer
criterion = nn.MSELoss()  # For regression problems, MSE is a good loss function
optimizer = optim.Adam(model.parameters(), lr=0.0001)  # Reduced learning rate
```

This cell contains the main training loop for the Wide & Deep model:

The model is set to training mode (model.train()).

For each epoch, the model iterates through batches of data provided by the train_loader. During each iteration:

    The model performs a forward pass where it predicts the ratings based on the input user and anime features.

    The MSE loss between the predicted and actual ratings is computed.

    The gradients are calculated through backpropagation using loss.backward(), and the model parameters are updated using optimizer.step().

    The running loss is accumulated to monitor the training progress.

The average loss per epoch is printed for tracking convergence. After the specified number of epochs, the training process is completed, and the model is ready for evaluation.

```python
# Training Loop
def train_model(model, train_loader, criterion, optimizer, num_epochs):
    model.train()
    for epoch in range(num_epochs):
        running_loss = 0.0
        for user_feats, anime_feats, ratings in train_loader:
            optimizer.zero_grad()
            outputs = model(user_feats, anime_feats)
            loss = criterion(outputs, ratings)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        
        print(f"Epoch {epoch+1}/{num_epochs}, Loss: {running_loss / len(train_loader)}")
    print("Training complete.")
```

In this cell, the evaluation process for the trained model is defined:

    The model is set to evaluation mode (model.eval()), which disables dropout and other training-specific behaviors.
    
    The function loops through batches of test data and makes predictions without updating the model’s parameters (since torch.no_grad() is used to prevent gradient computation).
    
    The loss is computed for the test data, allowing us to monitor how well the model generalizes to unseen data. Both the actual and predicted ratings are collected for further analysis.
   
    The function returns the actual ratings and predicted ratings, which can be used for plotting and calculating additional evaluation metrics (such as MAE, MSE, and R-squared).

This evaluation step ensures that the model’s performance can be assessed on unseen data, providing a reliable measure of its predictive power.

```python
# Evaluation function
def evaluate_model(model, test_loader, criterion):
    model.eval()
    actuals = []
    predictions = []
    total_loss = 0.0
    with torch.no_grad():
        for user_feats, anime_feats, ratings in test_loader:
            outputs = model(user_feats, anime_feats)
            loss = criterion(outputs, ratings)
            total_loss += loss.item()
            actuals.append(ratings.numpy())
            predictions.append(outputs.numpy())
    
    actuals = np.concatenate(actuals)
    predictions = np.concatenate(predictions)
    print(f"Test Loss: {total_loss / len(test_loader)}")
    
    # Return actual and predicted values for further analysis
    return actuals, predictions
```

In this cell, the r2_score function is defined to compute the R-squared metric, which measures how well the model's predictions align with the actual values. R-squared quantifies the proportion of the variance in the dependent variable (ratings) that is predictable from the independent variables (user and anime features). It ranges from 0 to 1, with values closer to 1 indicating better fit.

```python
# Calculating R-squared
def r2_score(actuals, predictions):
    ss_total = np.sum((actuals - np.mean(actuals))**2)
    ss_res = np.sum((actuals - predictions)**2)
    r2 = 1 - (ss_res / ss_total)
    return r2
```

This cell calculates three key evaluation metrics:

Mean Absolute Error (MAE): The average absolute difference between actual and predicted ratings.

Mean Squared Error (MSE): The average of the squared differences between actual and predicted ratings (used during training and evaluation).

R-squared (R²): Indicates the proportion of variance in the actual ratings explained by the model.

```python
# Calculating evaluation metrics
mae = np.mean(np.abs(actuals - predictions))
mse = np.mean((actuals - predictions)**2)
r2 = r2_score(actuals, predictions)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared: {r2}")
```

The MAE represents the average magnitude of the error in the model’s predictions, without considering whether the error is positive or negative. In this case, the MAE of approximately 0.974 suggests that, on average, the model’s predictions are about 1 rating point off from the true ratings. This is relatively acceptable for a recommendation system dealing with ratings that range from 1 to 10. However, the error of nearly 1 full point indicates the model could benefit from further tuning or architectural changes to improve its accuracy.

MSE measures the average squared difference between the actual and predicted ratings. Because it squares the error, larger deviations from the true values are penalized more heavily than smaller ones. With an MSE of 1.649, this suggests that while most predictions are close to the actual ratings, there are some significant outliers where the model’s predictions deviate substantially from the true values. While the MSE is in the same range as the MAE, the squared nature of this metric highlights larger errors more starkly. In practical terms, the MSE indicates the model has room to improve its handling of more extreme rating predictions.

R² (coefficient of determination) measures the proportion of variance in the actual ratings that the model can explain. An R² value of 0.387 means that the model explains about 38.7% of the variability in the ratings. This indicates that while the model has learned some useful patterns from the data, over 60% of the variability in the actual ratings is still unexplained. This suggests that the model’s predictive power is moderate but not particularly strong. Improving this score would involve optimizing the model architecture or exploring additional features that might better explain the variance in the data.

    MSE of 1.649: While most predictions are relatively close to the actual ratings, the presence of outliers with larger deviations is evident.
